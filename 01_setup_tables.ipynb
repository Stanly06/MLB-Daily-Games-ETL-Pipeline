```python

# 01 - Setup & Environment Initialization
# MLB Daily Games ETL Pipeline
# Create Database, Tables and Logging Structure


spark.sql("CREATE DATABASE IF NOT EXISTS mlb_project")

spark.sql("""
CREATE TABLE IF NOT EXISTS mlb_project.bronze_mlb_schedule_raw (
    execution_date STRING,
    source_api STRING,
    request_url STRING,
    status_code INT,
    raw_json STRING,
    extracted_at TIMESTAMP,
    pipeline_run_id STRING
)
USING DELTA
""")

spark.sql("""
CREATE TABLE IF NOT EXISTS mlb_project.silver_mlb_games (
    game_pk BIGINT,
    game_date STRING,
    official_date STRING,
    season STRING,
    game_type STRING,
    abstract_game_state STRING,
    game_status STRING,
    home_team_id INT,
    home_team_name STRING,
    away_team_id INT,
    away_team_name STRING,
    home_score INT,
    away_score INT,
    venue_id INT,
    venue_name STRING,
    current_inning INT,
    inning_state STRING,
    is_final BOOLEAN,
    source_api STRING,
    pipeline_run_id STRING,
    extracted_at TIMESTAMP,
    updated_at TIMESTAMP
)
USING DELTA
""")

spark.sql("""
CREATE TABLE IF NOT EXISTS mlb_project.gold_mlb_daily_summary (
    game_date STRING,
    total_games BIGINT,
    final_games BIGINT,
    non_final_games BIGINT,
    total_runs BIGINT,
    avg_runs_per_game DOUBLE,
    teams_playing BIGINT,
    last_updated_at TIMESTAMP,
    pipeline_run_id STRING
)
USING DELTA
""")

spark.sql("""
CREATE TABLE IF NOT EXISTS mlb_project.pipeline_logs (
    pipeline_name STRING,
    pipeline_run_id STRING,
    layer STRING,
    status STRING,
    message STRING,
    execution_date STRING,
    records_processed INT,
    started_at TIMESTAMP,
    ended_at TIMESTAMP,
    duration_seconds DOUBLE
)
USING DELTA
""")
```
